```python
map_center = [sum(New_Surrey_Kent_df['lat']) / len(New_Surrey_Kent_df['lat']), 
              sum(New_Surrey_Kent_df['long']) / len(New_Surrey_Kent_df['long'])]

# Create a dictionary of locations using DataFrame rows
locations = New_Surrey_Kent_df.to_dict('records')  # This converts rows to list of dicts
   
# Initialize the map
site_map = folium.Map(location=map_center, zoom_start=5)

# Loop through each location to add markers and circles to the map
for site in locations:
    town = site['Town']  # Get the town's name from the current site
    coords = (site['lat'], site['long'])  # Access coordinates from each record
    
    # Create a circle for the town location
    circle = folium.Circle(
        location=coords,
        radius=1000,
        color='#d35400',
        fill=True
    ).add_child(folium.Popup(f'Town: {town}'))  # Popup shows town name
    
    # Create a marker for the town location
    marker = folium.Marker(
        location=coords,
        icon=folium.DivIcon(  # Use folium.DivIcon instead of DivIcon (just a correction)
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{town}</b></div>'  # Display town name in marker
        )
    )
    
    # Add the circle and marker to the map
    site_map.add_child(circle)
    site_map.add_child(marker)

# Display the map
site_map
```

```python
def marker_color(result):
    return 'green' if result >= 70 else 'red'

# Apply the marker_color function to create a new column 'marker_color'
New_Surrey_Kent_df['marker_color'] = New_Surrey_Kent_df['school_score_1'].apply(marker_color)

# Initialize the folium map
site_map = folium.Map(location=[51.0, -0.2], zoom_start=10)  

# Create marker cluster
marker_cluster = MarkerCluster()
site_map.add_child(marker_cluster)

# Add markers to the cluster
for _, row in New_Surrey_Kent_df.iterrows():
    coordinate = [row['lat'], row['long']]
    folium.Marker(
        location=coordinate,
        icon=folium.Icon(color=row['marker_color'], icon_color=row['marker_color'])
    ).add_to(marker_cluster)

# Display the map
site_map
```
